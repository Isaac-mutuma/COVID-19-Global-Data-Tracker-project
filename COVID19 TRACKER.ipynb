```python
import pandas as pd

# Load dataset
df = pd.read_csv("owid-covid-data.csv")

# Convert date column to datetime
df['date'] = pd.to_datetime(df['date'])

# View the structure
df.info()
df.head()
```

```python
# Filter for selected countries
countries = ['Kenya', 'United States', 'India']
df = df[df['location'].isin(countries)]

# Drop rows with missing key values
df = df.dropna(subset=['total_cases', 'total_deaths', 'total_vaccinations'])

# Fill the remaining missing numeric values
df.fillna(0, inplace=True)
```

```python
import matplotlib.pyplot as plt

for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.grid(True)
plt.show()
```

```python
for country in countries:
    country_data = df[df['location'] == country]
    plt.plot(country_data['date'], country_data['total_vaccinations'], label=country)

plt.title("Total Vaccinations Over Time")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend()
plt.grid(True)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt


# Load the dataset
df = pd.read_csv('owid-covid-data.csv')

# Preview the first 5 rows
df.head()
```

```python
# View the shape and column names
print("Shape:", df.shape)
print("\nColumns:\n", df.columns)
```

```python
# Check for missing values
df.isnull().sum().sort_values(ascending=False)
```

```python
# Summary statistics
df.describe()
```

```python
# Summary statistics
df.describe()
# Check the size of the dataset and available columns
print("Shape of dataset:", df.shape)
print("\nAvailable columns:\n", df.columns.tolist())
```

```python
# See how many missing values are in each column
missing_values = df.isnull().sum()
missing_values = missing_values[missing_values > 0].sort_values(ascending=False)

print("Missing values in each column:\n")
print(missing_values)
```

```python
# Drop columns with more than 50% missing values
threshold = len(df) * 0.5
df_cleaned = df.dropna(thresh=threshold, axis=1)

# Drop rows with missing values in crucial columns
df_cleaned = df_cleaned.dropna(subset=['total_cases', 'total_deaths', 'location', 'date'])

print("Cleaned dataset shape:", df_cleaned.shape)
```

```python
import matplotlib.pyplot as plt

# Filter for a country, e.g., Kenya
country = 'Kenya'
country_df = df_cleaned[df_cleaned['location'] == country]

plt.figure(figsize=(12, 6))
plt.plot(country_df['date'], country_df['total_cases'], marker='o', linestyle='-', color='blue')
plt.title(f"Total COVID-19 Cases Over Time in {country}")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Choose countries to compare
countries = ['Kenya', 'India', 'United States']

plt.figure(figsize=(14, 7))

for country in countries:
    country_df = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_df['date'], country_df['total_cases'], label=country)

plt.title("Total COVID-19 Cases Over Time")
plt.xlabel("Date")
plt.ylabel("Total Cases")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Daily new cases
plt.figure(figsize=(14, 7))

for country in countries:
    country_df = df_cleaned[df_cleaned['location'] == country]
    plt.plot(country_df['date'], country_df['new_cases'], label=country)

plt.title("Daily New COVID-19 Cases")
plt.xlabel("Date")
plt.ylabel("New Cases")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```

```python
# Step: Plot Total Vaccinations Over Time
plt.figure(figsize=(14, 7))

for country in countries:
    country_df = df_cleaned[df_cleaned['location'] == country]
    
    plt.plot(country_df['date'], country_df['total_vaccinations'], label=country)

plt.title("Total COVID-19 Vaccinations Over Time")
plt.xlabel("Date")
plt.ylabel("Total Vaccinations")
plt.legend()
plt.xticks(rotation=45)
plt.grid(True)
plt.tight_layout()
plt.show()
```
